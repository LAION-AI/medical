```python
import wandb
api = wandb.Api()

# run is specified by <entity>/<project>/<run id>
run = api.run("jimgoo/laion-fmri/zskzyi2m")

# save the metrics for the run to a csv file
df = run.history()
# metrics_dataframe.to_csv("metrics.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
df2 = df[['train/loss_prior', 'train/loss_nce']]
df2.plot()
```

```python
df2.head(3).plot();
```

```python
df2.head()
```

```python
# default currently
epochs = len(df2)
alpha = np.ones(epochs) * 0.01

df3 = df2.copy()
df3.iloc[:,0] *= alpha
df3.iloc[:,1] *= (1-alpha)
df3.plot();
```

```python
(df3.iloc[:,0] / df3.iloc[:,1]).plot();
```

```python
epochs = len(df2)
#alpha = np.ones(epochs) * 0.01
alpha = np.linspace(0.01, 0.05, epochs, endpoint=True)

df3 = df2.copy()
df3.iloc[:,0] *= alpha
df3.iloc[:,1] *= (1-alpha)

df3.plot();
```

```python
(df3.iloc[:,0] / df3.iloc[:,1]).plot();
```
